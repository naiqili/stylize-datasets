```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# mu

```python
mu = np.zeros((10000, 960))

fo = open('mu_stats.txt');

x = 0
for l in fo:
    t = l.strip().split()
    for y in range(len(t)):
        mu[x][y] = float(t[y])
    x += 1

fo.close()
```

```python
plt.hist(mu[:, 0], bins=20)
```

```python
plt.hist(mu[:, 100], bins=20)
```

```python
plt.hist(mu[:, 200], bins=20)
```

```python
plt.hist(mu[:, 900], bins=20)
```
